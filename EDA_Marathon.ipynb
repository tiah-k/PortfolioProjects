```python
#Clean up Athlete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
# Remove "h" from athlete performance 
```

```python
#Revive the column deleted values from the original DataFrame 
df2['Athlete performance'] = df['Athlete performance']
```

```python
df2 = df2.drop(['Athlete club','Athlete country', 'Athlete year of birth', 'Athlete age category' ], axis = 1)
```

```python
df2.isna().sum()
```

```python
df2['athlete-age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
# Rename columns
```

```python
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#athlete_age                  float64
#Athlete perfomrance           object
#athlete-age                    int32
```

```python
df2 = df2.rename(columns = {'Year of event':'year' , 
                            'Event dates' : 'race_day', 
                            'Event name' : 'race_name',
                            'Event distance/length': 'race_length', 
                            'Event number of finishers' : 'race_number_of_fininshers', 
                            'Athlete performance' : 'athlete_performance', 
                            'Athlete gender' : 'athlete_gender', 
                            'Athlete average speed' : 'athlete_average_speed', 
                            'Athlete ID' : 'athlete_id' 
                    
                })
df2 = df2.drop(['Athlete perfomrance'], axis = 1)
```

```python
# Questions to answer from the dataset 
```

```python
# Differrence in speed for the 50k, 50mi male to female
```

```python
df2.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# What age groups are the best in the 50m race (20+ races min)
```

```python
df2.query('race_length =="50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19')
```

```python
# What age groups are the worst  in the 50m race (20+ races min)
```

```python
df2.query('race_length =="50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>19')
```
